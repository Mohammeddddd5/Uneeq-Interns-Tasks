```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_excel("UberDataset.xlsx")
df.head()
```

```python
df.info()
```

```python
pd.set_option('display.max_rows', None)

df.head(60)
```

```python
for col in ["START_DATE","END_DATE"]:
    df[col] = pd.to_datetime(df[col], errors='coerce', infer_datetime_format=True).dt.strftime('%Y-%m-%d %H:%M:%S')

df.head(60)
```

```python
df["START_TIME"] = pd.to_datetime(df["START_DATE"]).dt.time
df["END_TIME"] = pd.to_datetime(df["END_DATE"]).dt.time
df["START_DATE"] = pd.to_datetime(df["START_DATE"]).dt.date
df["END_DATE"] = pd.to_datetime(df["END_DATE"]).dt.date
df.head(60)
```

```python
df.info()
```

```python
df.isna().sum() / len(df) * 100
```

```python
sns.countplot(
    data=df,
    x="PURPOSE",
    order=df["PURPOSE"].value_counts().index
)

plt.xticks(rotation=45)
plt.show()
```

```python
df['PURPOSE'].isna().mean()  # check % missing
df.groupby(["CATEGORY"])['PURPOSE'].apply(lambda x: x.isna().mean())
```

```python
df.loc[df["PURPOSE"].isna() & (df["CATEGORY"] == "Business"), "PURPOSE"] = "Business Travel"
df.loc[df["PURPOSE"].isna() & (df["CATEGORY"] == "Personal"), "PURPOSE"] = "Personal Travel"
df.groupby(["CATEGORY"])['PURPOSE'].apply(lambda x: x.isna().mean())
```
